###Unir los archivos de protocolos para generar uno solo

```python
def comparar_y_agregar_registros(archivo1, archivo2):
    try:
        # Cargar los archivos CSV en DataFrames de pandas
        df_completo = archivo1

        # Identificar los registros del archivo2 que no están en el archivo1 por el ID del TT
        registros_nuevos = archivo2[~archivo2['TT'].isin(archivo1['TT'])]

        # Agregar los nuevos registros al DataFrame completo
        df_completo = pd.concat([df_completo, registros_nuevos], ignore_index=True, sort=False)

        # Guardar el DataFrame actualizado en un nuevo archivo CSV
        df_completo.to_csv('/content/drive/MyDrive/Datasets/Protocolos_completo.csv', index=False)

        print("Registros agregados con éxito.")

    except FileNotFoundError:
        print("Error: Uno o ambos archivos CSV no se encontraron.")
    except KeyError:
        print("Error: La columna 'TT' no existe en uno o ambos archivos CSV.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
```

```python
def actualizar_sinodales(Protocolos_completo, sinodales):
    try:
        # Iterar sobre las filas del DataFrame de los protocolos
        for index, row in Protocolos_completo.iterrows():
            tt = row['TT']
            if pd.isna(row['sinodal1']) or pd.isna(row['sinodal2']) or pd.isna(row['sinodal3']):
                # Buscar la fila correspondiente en el DataFrame de los sinodales
                fila_sinodal = sinodales[sinodales['TT'] == tt]
                if not fila_sinodal.empty:
                    # Actualizar los valores NaN en las columnas sinodales (archivo de protocolos)
                    if pd.isna(row['sinodal1']):
                        Protocolos_completo.loc[index, 'sinodal1'] = sinodales['DOCENTE 3'].values[0]
                    if pd.isna(row['sinodal2']):
                        Protocolos_completo.loc[index, 'sinodal2'] = sinodales['DOCENTE 4'].values[0]
                    if pd.isna(row['sinodal3']):
                        Protocolos_completo.loc[index, 'sinodal3'] = sinodales['DOCENTE 5'].values[0]

        # Guardar el DataFrame actualizado en el archivo CSV "completo"
        Protocolos_completo.to_csv('/content/drive/MyDrive/Datasets/Protocolos_completo.csv', index=False)

        print("Valores de sinodales actualizados con éxito.")

    except FileNotFoundError:
        print("Error: Uno o ambos archivos CSV no se encontraron.")
    except KeyError:
        print("Error: Una o más columnas necesarias no existen en los archivos CSV.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

#Realizar la actualización de datos
actualizar_sinodales(Protocolos_completo, sinodales)
```

```python
import re

def limpiar_objetivos(texto):
    if isinstance(texto, str):  # Verificar que el valor de "texto" sea un string

        texto = re.sub(r"\n", " ", texto).strip()  # Eliminar saltos de línea
        texto = re.sub(r"[\•\●\\▪\○\➢\·\\■\✓\\\\❖]", "", texto).strip()  # Limpiar viñetas
        texto = re.sub(r"\d+[.:]", "", texto).strip() # 2018-B082
        # Eliminar los numeros romanos que son usados como viñetas como: ii iii y tambien letras como: a. y b.  TT: 2021-B071
        texto = re.sub(r'(?<!["\w])\b(i|ii|iii|iv|v|vi|vii|viii|ix|x|a|b|c|d|e|f|g|h|i|j|k|A|C|D|E|F|G|H|I|J|K)\b(?![-\s,])', '', texto)
        # Ademas, evitamos eliminar letras entre comillas por ejemplo de: Sección "B".  TT: 2021-B069

        # ?<!\s|[-]) checa que no haya un espacio o guion detras del numero, para evitar borrar numeros en palabras como COVID-19. TT: 2021-B019
        # Tomamos un rango de edad entre 0 y 99 años para las edades y no eliminamos si el numero es seguido de ° o un año ( d{4} ).  TT: 2021-B059
        texto = re.sub(r"(?<!\s|[-])\b\d{1,2}(?:\.\d{1,2}){0,2}\b(?!°|º|\s?\d{4})", "", texto).strip()

        # Eliminar URL's.   TT: 2021-B004
        texto = re.sub(r"https?://\S+", "", texto)
        # Eliminar [] o [[ ]].  TT: 2021-B004
        texto = re.sub(r"[\[\]]", "", texto)
        # Quita codigos como "_x000C_".   TT: 2021-B007
        texto = re.sub(r"_x[0-9A-Za-z]+_", "", texto)
        # Borrar // ó – Ejemplo de TT con num: 	2021-B029 y 2021-B048
        texto = re.sub(r"\s?//\s?|\s?–\s?", " ", texto).strip()
        # Eliminar el parentesis cuando es usado como viñeta.   TT: 2021-B071
        texto = re.sub(r"\b\d+\)\s*|(?<![a-zA-Z0-9])\)", "", texto).strip()
        # Limpiar guiones usados como viñetas, se usa esta regex para evitar cortar guiones en palabras como COVID-19.  TT:2020-A107 y TT: 2021-B045
        texto = re.sub(r"(?<![a-zA-Z0-9])-", "", texto).strip()
        # Eliminar viñeta "o".  TT: 2021-B057 y TT: 2016-A027
        texto = re.sub(r"\b[oO]\s(?=[A-Z])", "", texto).strip()
        # eliminar un numero q aparezca al final del texto, como en el TT 2021-B066
        texto = re.sub(r"\s\d+$", "", texto).strip()
        # Eliminar número que tiene despues una palabra como "Específicos", etc.  TT: 2021-B056
        texto = re.sub(r"\b\d+\s*(?=(Particulares|Específicos|específicos|Objetivos|Objetivo|Generar|Implementar|General|Servidor|- )\b)", "", texto).strip()
        # Eliminar . ó : que tienen antes una palabra como "Específicos", etc.  TT: 2021-B006
        texto = re.sub(r"(Particulares|Específicos|específicos|Objetivos|Objetivo|general|General|particulares|objetivos|objetivo|Específico)[.:]\s*", r"\1 ", texto)
        # eliminar . ó : del inicio que parece que son usados como viñetas.   # TT: 2015-B066
        texto = re.sub(r"^[.:]+", "", texto).strip()
        # Quitar alguna letra que este al inicio y sola.  TT: 2016-A027
        texto = re.sub(r"^\s*[a-zA-Z]\b", "", texto).strip()
        # Quitar guiones pero respetando las posibles palabras con guiones intermedios, como e-learning.  TT: 2021-B014
        texto = re.sub(r"\s-\s", " ", texto).strip()
        # Eliminar i como viñeta. TT: 2016-A012
        texto = re.sub(r"\bi\s", "", texto)

    return texto
```

```python
def limpiar_palabras_clave(texto):
    if isinstance(texto, str):  # Verificar que el valor de "texto" sea un string
        # Eliminar saltos de línea
        texto = re.sub(r"\n", " ", texto).strip()
        # Limpiar viñetas
        texto = re.sub(r"[\•\●\\▪\○\➢\·\\■\✓\\\\–]", "", texto).strip()
        # Limpiar guiones usados como viñetas, se usa esta regex para evitar cortar guiones en palabras como COVID-19.   TT: 2021-B032
        texto = re.sub(r"(?<![a-zA-Z0-9])-", "", texto).strip()
        # Limpiar : y . usados como viñetas TT: 2021-B018, TT: 2021-B049  y TT: 2021-A072
        texto = re.sub(r"^[.:—]+", "", texto).strip()
        # Eliminar el 1 de digitales1 o en puntos (.) en Aplicación Móvil. 1.    TT:2021-B049, TT: 2021-B087 y TT: 2023-A088
        texto = re.sub(r"(?<=digitales|Digitales)\s?\d+\b|(?<=\.)\s*\d+\b", "", texto)
    return texto
```

```python
def limpiar_resumen(texto):
    if isinstance(texto, str):  # Verificar que el valor de "texto" sea un string
        # Eliminar saltos de línea
        texto = re.sub(r"\n", " ", texto).strip()
        # Limpiar viñetas TT: 2021-A009, TT: 2019-B109 (:), TT: 2021-A078 (.-) y TT: 2016-B102 (—)
        texto = re.sub(r"[–\:\.\–\—]", "", texto).strip()
        # Limpiar guiones usados como viñetas, se usa esta regex para evitar cortar guiones en palabras como COVID-19
        texto = re.sub(r"(?<![a-zA-Z0-9])-", "", texto).strip()
    return texto
```

```python
protocolos_limpios["objetivos"] = protocolos_limpios["objetivos"].apply(limpiar_objetivos)
protocolos_limpios["claves"] = protocolos_limpios["claves"].apply(limpiar_palabras_clave)
protocolos_limpios["resumen"] = protocolos_limpios["resumen"].apply(limpiar_resumen)
```

```python
def cargar_modelo(nombre_modelo):
    print(f"Cargando modelo: {nombre_modelo}")
    modelo = SentenceTransformer(nombre_modelo)
    return modelo

def obtener_embeddings(df, columna, modelo):
    textos = df[columna].astype(str).tolist()
    print(f"Obteniendo embeddings para la columna '{columna}'...")
    embeddings_obtenidos = modelo.encode(textos, convert_to_tensor=False)
    # .encode: convierte el texto (textos) en vectores numéricos (embeddings), esto solo funciona con SentenceTransformer
    # convert_to_tensor=False indica que el resultado se devolvera como un array de numpy en lugar de un tensor de pytorch
    return embeddings_obtenidos

def guardar_embeddings(embeddings, nombre_archivo):
    with open(nombre_archivo, "wb") as f:
        # usamos write binary para almacenar los pkl's
        pickle.dump(embeddings, f)
        # con dump guardamos los embeddings en el archivo que llamamos "f"
    print(f"Embeddings guardados en {nombre_archivo}")

def main():
    try:
        df = pd.read_csv("protocolos_completo_limpios.csv")
    except Exception as e:
        print("Error al leer el archivo.", e)
        return

    # campo a comparar
    campo = input("Selecciona el campo (Titulo, resumen, objetivos, claves) para obtener sus PKL's: ").strip()
    # si se ingresa mal escrito el campo o no está en el df, entonces:
    if campo not in df.columns:
        print("El campo ingresado no se encuentra en los datos.")
        return

    # ruta para guardar los pkl's
    ruta_destino = "C:/Users/1010h/Desktop/Embeddings/Embeddings/PKL"

    # iterar sobre cada modelo para calcular y guardar los embeddings
    for clave_modelo, nombre_modelo in MODELOS.items():
        modelo = cargar_modelo(nombre_modelo)
        embeddings_obtenidos = obtener_embeddings(df, campo, modelo)
        archivo_embeddings = os.path.join(ruta_destino, f"{campo}_{clave_modelo}_embeddings.pkl")
        guardar_embeddings(embeddings_obtenidos, archivo_embeddings)

if __name__ == "__main__":
    main()

############
# top modelos:
# EL mejor es el 3, luego el 1, 2 y al ultimo el 4
```

```python
# Solicitar consulta al usuario
while True:
    query = input("Ingresa el texto para comparar: ").strip()
    if not query:
        print("Consulta vacía. Por favor, ingresa un texto.")
        continue
    print("\nConsulta: ", query)
    break

# Lista para guardar nuevos resultados
filas_tabla = []

# Comparar con cada modelo
for clave_modelo, nombre_modelo in MODELOS.items():
    print(f"Procesando modelo: {clave_modelo}")
    modelo = SentenceTransformer(nombre_modelo)
    df_temp = df.copy()

    for campo in CAMPOS:
        nombre_pkl = f"{campo}_{clave_modelo}_embeddings.pkl"
        ruta_pkl = os.path.join("PKL", nombre_pkl)

        if not os.path.exists(ruta_pkl):
            print(f"No se encontró el archivo {ruta_pkl}")
            continue

        with open(ruta_pkl, "rb") as f:
            embeddings_cargados = pickle.load(f)

        embedding_query = modelo.encode(query, convert_to_tensor=False)
        sim_coseno = cosine_similarity([embedding_query], embeddings_cargados)[0]
        df_temp[f"similitud_{campo}"] = sim_coseno

    columnas_sim = [f"similitud_{c}" for c in CAMPOS if f"similitud_{c}" in df_temp.columns]
    if not columnas_sim:
        continue

    df_temp["similitud_total"] = df_temp[columnas_sim].sum(axis=1)
    df_top = df_temp.sort_values("similitud_total", ascending=False).head(5)

    for i, (_, fila) in enumerate(df_top.iterrows(), start=1):
        for campo in CAMPOS:
            texto = fila[campo]
            fila_resultado = {
                "Modelo": clave_modelo,
                "Consulta": query if campo == "Titulo" and i == 1 else "",
                "No. Resultado más parecido": i,
                "Campo": campo,
                "Texto": texto,
                "Similitud": fila.get(f"similitud_" + campo, 0),
                "Justificación": ""
            }
            filas_tabla.append(fila_resultado)

        filas_tabla.append({
            "Modelo": clave_modelo,
            "Consulta": "",
            "No. Consulta": i,
            "Campo": "TOTAL",
            "Texto": "",
            "Similitud": fila["similitud_total"],
            "Justificación": ""
        })

# Crear DataFrame con nuevos resultados
df_nuevos = pd.DataFrame(filas_tabla)

# Eliminar duplicados de nombres de columnas si los hay
df_nuevos = df_nuevos.loc[:, ~df_nuevos.columns.duplicated()]

# Nombre del archivo de salida
nombre_archivo = "tabla_consultas_formato_deseado.csv"

# Cargar datos existentes si el archivo ya existe
if os.path.exists(nombre_archivo):
    df_existente = pd.read_csv(nombre_archivo)
    # Asegurar mismo orden de columnas al unir
    df_nuevos = df_nuevos[df_existente.columns]
    df_final = pd.concat([df_existente, df_nuevos], ignore_index=True)
else:
    df_final = df_nuevos

# Guardar resultados finales
df_final.to_csv(nombre_archivo, index=False)
print(f"✅ Resultados guardados en '{nombre_archivo}' sin columnas duplicadas.")
```

```python
import pandas as pd
import re
from transformers import AutoModelForTokenClassification, TrainingArguments, Trainer, AutoTokenizer, pipeline, DataCollatorForTokenClassification
from datasets import load_dataset, Dataset, concatenate_datasets
from huggingface_hub import login, HfApi, upload_folder
```

###XML-ROBERTA-LARGE

```python
modelo = AutoModelForTokenClassification.from_pretrained("iEsmeralda/ner_xlm_roberta_tech")
tokenizer = AutoTokenizer.from_pretrained("iEsmeralda/ner_xlm_roberta_tech")
ner_pipeline = pipeline("ner", model=modelo, tokenizer=tokenizer, aggregation_strategy="simple")

texto = "El proyecto de bases de datos fue desarrollado por los estudiantes, ademas usaron redes neuronales y algo de visión por computadora y Gabriel García Márquez era buen escritor"
resultados = ner_pipeline(texto)

for r in resultados:
    print(f"Entidad: {r['word']} - Tipo: {r['entity_group']} - Score: {r['score']:.3f}")
```

```python
# agrupar entidades consecutivas si son del mismo tipo y están juntas
def agrupar_entidades_consecutivas(entidades):
    if not entidades:
        return []

    entidades_agrupadas = []
    entidad_actual = entidades[0].copy()

    for i in range(1, len(entidades)):
        entidad = entidades[i]

        # Si la entidad es del mismo tipo y esta justo despues de la anterior, entonces:
        if entidad["entity_group"] == entidad_actual["entity_group"] and entidad["start"] == entidad_actual["end"] + 1:
            # Se concatena la palabra y se actualiza el rango y se toma un score promedio de ambas
            entidad_actual["word"] += " " + entidad["word"]
            entidad_actual["end"] = entidad["end"]
            entidad_actual["score"] = (entidad_actual["score"] + entidad["score"]) / 2
        else:
            entidades_agrupadas.append(entidad_actual)
            entidad_actual = entidad.copy()

    entidades_agrupadas.append(entidad_actual)
    return entidades_agrupadas
```

```python
# Cargar el DataFrame
pruebas = pd.read_csv('/content/Pruebas modelos NER casi completado.csv')

modelo = AutoModelForTokenClassification.from_pretrained("iEsmeralda/ner_xlm_roberta_tech")
tokenizer = AutoTokenizer.from_pretrained("iEsmeralda/ner_xlm_roberta_tech")
ner_pipeline = pipeline("ner", model=modelo, tokenizer=tokenizer, aggregation_strategy="simple")

# Resultados para este modelo
ner_resultados = []
accuracy_resultados = []

for idx, fila in pruebas.iterrows():
    if fila['Modelo'] == 'XML-ROBERTA-LARGE':
        texto = str(fila['Contenido'])
        entidades = ner_pipeline(texto)
        entidades_agrupadas = agrupar_entidades_consecutivas(entidades)
        entidades_modelo = [ent['word'] for ent in entidades_agrupadas]
        ner_resultados.append(', '.join(entidades_modelo))

        # Calcular accuracy
        entidades_manual = str(fila['NER encontradas manualmente']).lower().split(', ')
        entidades_modelo_limpias = [e.lower() for e in entidades_modelo]
        if entidades_modelo_limpias:
            aciertos = sum(1 for e in entidades_modelo_limpias if e in entidades_manual)
            accuracy = aciertos / len(entidades_modelo_limpias)
        else:
            accuracy = 0.0

        accuracy_resultados.append(round(accuracy, 2))
    else:
        ner_resultados.append(fila['NER del modelo'])  # mantener valores originales
        accuracy_resultados.append(fila['Accuracy (#NER modelo entre #NER manuales)'])

# Actualizar DataFrame
pruebas['NER del modelo'] = ner_resultados
pruebas['Accuracy (#NER modelo entre #NER manuales)'] = accuracy_resultados

# Guardar archivo CSV
pruebas.to_csv('/content/Pruebas modelos NER completado.csv', index=False)
```
